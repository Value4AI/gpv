```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

plt.rcParams['font.family'] = 'Arial Unicode MS'


# 数据
data = {
    "唐僧": {
        "Universalism": 0.8920416269037459,
        "Hedonism": 0,
        "Achievement": 0.9876969986491733,
        "Power": 0.7332806076322284,
        "Security": 0.9375,
        "Self-Direction": -0.3240296407179399,
        "Stimulation": 0,
        "Tradition": 0.9588095699295853,
        "Benevolence": 1,
        "Conformity": -0.13770889987548193
    },
    "悟空": {
        "Universalism": 1,
        "Hedonism": 0,
        "Achievement": 1,
        "Power": 0.974189281463623,
        "Security": 0.7333333333333333,
        "Self-Direction": 0.5384615384615384,
        "Stimulation": 1,
        "Tradition": 1,
        "Benevolence": 0.875,
        "Conformity": 0.5
    },
    "八戒": {
        "Universalism": 0,
        "Hedonism": 0.8,
        "Achievement": 0.8433399399121603,
        "Power": 0.9719934649765491,
        "Security": 0.9435934917284892,
        "Self-Direction": -0.08650501403543684,
        "Stimulation": 0,
        "Tradition": 0.6305661116327558,
        "Benevolence": 0.8316367268562317,
        "Conformity": 0.23076923076923078
    },
    "沙僧": {
        "Universalism": 1,
        "Hedonism": 0,
        "Achievement": 1,
        "Power": 0,
        "Security": 1,
        "Self-Direction": -0.4151780651165889,
        "Stimulation": 0,
        "Tradition": 0.9016852378845215,
        "Benevolence": 0.7267772005154536,
        "Conformity": 0.7003267755111059
    }
}

# 转换为列表并提取维度
labels = list(data["唐僧"].keys())
num_vars = len(labels)

# 定义角度，用于雷达图的坐标
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# 数据处理
characters = ["唐僧", "悟空", "八戒", "沙僧"]
data_values = {name: [v for v in data[name].values()] for name in characters}

# 为了雷达图闭合最后一个点
for name in characters:
    data_values[name].append(data_values[name][0])
angles.append(angles[0])

# 创建雷达图
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# 绘制每个角色的雷达图
for name in characters:
    ax.fill(angles, data_values[name], alpha=0.25, label=name)
    ax.plot(angles, data_values[name], linewidth=2)

# 设置图例和标签
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels, fontsize=15)

plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.2), fontsize=15)
plt.title('价值测定结果', size=20, y=1.1)
plt.show()
```
